# Introducción

El **Fondo de Aportaciones para el Fortalecimiento de los Municipios y de las Demarcaciones Territoriales del Distrito Federal**^[[Página oficial FORTAMUN](https://www.gob.mx/sesnsp/acciones-y-programas/fondo-de-aportaciones-para-el-fortalecimiento-de-los-municipios-y-de-las-demarcaciones-territoriales-del-distrito-federal-fortamun)] (FORTAMUN) tiene como propósito fortalecer las capacidades administrativas y financieras de los gobiernos municipales, así como apoyar el cumplimiento de sus obligaciones constitucionales.
Si bien el FORTAMUN es un fondo de carácter amplio, su marco normativo establece que una proporción de los recursos puede destinarse a la atención de necesidades directamente vinculadas con la seguridad pública, particularmente en áreas estratégicas como el fortalecimiento del estado de fuerza, equipamiento, profesionalización y demás acciones alineadas a los Programas con Prioridad Nacional. 

En este contexto, y en cumplimiento de lo dispuesto por el Presupuesto de Egresos de la Federación 2023, el Consejo Nacional de Seguridad Pública emitió el **Exhorto 02/XLVIII/22**^[[Acuerdos y Exhortos del Consejo Nacional de Seguridad Pública](https://www.dof.gob.mx/nota_detalle_popup.php?codigo=5675407)], mediante el cual se conmina a los municipios y a las demarcaciones territoriales de la Ciudad de México a destinar al menos el veinte por ciento de los recursos a necesidades directamente vinculadas con la seguridad pública.
Este exhorto constituye un mecanismo de articulación y alineación de la política pública municipal con las prioridades nacionales en la materia, y orienta el uso eficiente y estratégico de los recursos federales.

Con base en los *Criterios para la Coordinación de la aplicación y verificación de los recursos del FORTAMUN en materia de seguridad pública para el ejercicio fiscal 2024*, la Dirección General de Planeación del Secretariado Ejecutivo del Sistema Nacional de Seguridad Pública elabora el presente Informe sobre el ejercicio de los recursos del FORTAMUN destinados a seguridad pública.
Para ello, se utilizan los Informes de Conclusión remitidos por los municipios, a través de los Secretariados Ejecutivos Estatales, y compilados por la Dirección General de Vinculación y Seguimiento como ventanilla única.

El **Informe FORTAMUN 2024** presenta un resumen del cumplimiento en la entrega de informes municipales y ofrece un análisis integral de cuatro elementos clave: 

1. El grado de cumplimiento al Exhorto 02/XLVIII/22 respecto a la proporción de recursos destinados a seguridad pública
2. La inversión, adquisiciones y población beneficiaria por subprograma
3. La información reportada sobre el estado de fuerza municipal
4. Las retenciones

Este documento contribuye a fortalecer la transparencia, la rendición de cuentas y la toma de decisiones basada en evidencia en el uso de los recursos federales destinados a la seguridad pública municipal.

# Metodología

El presente Informe se elaboró con base en la información contenida en los Informes de Conclusión remitidos por los municipios y demarcaciones territoriales, los cuales fueron integrados y enviados al Secretariado Ejecutivo del Sistema Nacional de Seguridad Pública (SESNSP) a través de los Secretariados Ejecutivos Estatales.
La Dirección General de Vinculación y Seguimiento (DGVyS) fungió como ventanilla única responsable de la recepción, revisión preliminar y resguardo de dichos documentos, conforme a lo dispuesto en los Criterios FORTAMUN 2024. 

La Dirección General de Planeación (DGP) consolidó y analizó la información disponible a partir de los insumos municipales, con el objetivo de evaluar el ejercicio de los recursos del FORTAMUN destinados a la seguridad pública, así como el grado de cumplimiento al Exhorto 02/XLVIII/22 del Consejo Nacional de Seguridad Pública. 

La información utilizada presenta diversas limitaciones inherentes al proceso de captura, integración y reporte, entre las cuales destacan: 

- Variabilidad en la calidad de los Informes de Conclusión, debido a diferencias en capacidades técnicas municipales. Esto resulta en la ausencia de un estándar uniforme para la presentación y el análisis de la información. 

- Inconsistencias o vacíos en desagregaciones solicitadas (por ejemplo, beneficiarios, estado de fuerza o detalles por subprograma). 

- Errores de captura, formatos no estandarizados o discrepancias entre anexos dentro de un mismo informe municipal, que afectan la veracidad de la información presentada en los informes.

# Cumplimiento al Exhorto 02/XLVIII/22

El Exhorto 02/XLVIII/22, aprobado por el Consejo Nacional de Seguridad Pública en su Cuadragésima Octava Sesión Ordinaria celebrada el 2 de diciembre de 2022 y publicado en el Diario Oficial de la Federación el 23 de diciembre de 2022, establece que los Municipios y las Demarcaciones Territoriales de la Ciudad de México deberán destinar, por cada ministración que reciban del FORTAMUN —con fundamento en los artículos 37 y 38 de la Ley de Coordinación Fiscal—, al menos el 20% de dichos recursos a necesidades directamente vinculadas con la seguridad pública, en congruencia con los propósitos de los Programas y Subprogramas con Prioridad Nacional vigentes. 

Para garantizar el cumplimiento del exhorto, el Consejo instruyó al SESNSP verifique la correcta aplicación del porcentaje mínimo obligatorio destinado a seguridad pública por parte de los municipios. 

- Porcentaje de municipios que cumplieron por estado. 
- Porcentaje a nivel nacional.  
- Gráfica de barras con porcentaje de cumplimiento por estado y del porcentaje de cumplimiento promedio nacional.

Para evaluar el grado de alineación de los municipios con el Exhorto 02/XLVIII/22 del Consejo Nacional de Seguridad Pública —que conmina a destinar al menos el 20% de los recursos del FORTAMUN a acciones directamente vinculadas con la seguridad pública— se analizó el porcentaje de municipios que reportaron haber cumplido con este criterio.
El siguiente gráfico muestra la proporción de cumplimiento por entidad federativa, lo que permite vislumbrar diferencias territoriales significativas en la orientación del gasto municipal hacia la seguridad pública y en las capacidades locales para planear, clasificar y documentar el uso de los recursos del Fondo.

```python
# libraries
import numpy as np
import pandas as pd
import polars as pl
import duckdb as db
from pathlib import Path
import plotly.express as px
from plotnine import *
import warnings
warnings.filterwarnings('ignore')
```

```python
# database file path 
database = Path('../../data_lake/3-golden/dae.duckdb')
```

```python
# database connection
conn = db.connect(database)
```

```python
porcentaje = conn.sql('''
    select Entidad_Federativa,
        Municipio,
        round(coalesce("Seg_pub%", 0), 2) porcentaje
    from fortamun_resumen_2024
''').pl()
```

```python
# 2. Definir la variable de cumplimiento (para el color del punto)
# Usando la sintaxis de Polars para crear la nueva columna
porcentaje = porcentaje.with_columns(
    (pl.when(pl.col('porcentaje') >= 0.20)
        .then(pl.lit("Sí cumplió"))
        .otherwise(pl.lit("No cumplió"))
    .alias('cumplio_str') # Esta será la nueva columna usada para el color
    ),
    (pl.int_range(1, pl.len() + 1).alias('mun_id')),
)

# 3. Crear el gráfico con plotnine
referencia = 0.20  # La meta de cumplimiento del 20%
```

```python
cumplimiento = porcentaje.clone()
```

```python
cumplio_estado = (
    cumplimiento
        .with_columns(
            # Conteo de municipios que 'Sí cumplió' por cada Entidad Federativa
            pl.col("cumplio_str")
            .eq("Sí cumplió")
            .sum()
            .over("Entidad_Federativa")
            .alias("total_si_cumplio_estado"),

            # Conteo de municipios que 'No cumplió' por cada Entidad Federativa
            pl.col("cumplio_str")
            .eq("No cumplió")
            .sum()
            .over("Entidad_Federativa")
            .alias("total_no_cumplio_estado"),
            )
        .select('Entidad_Federativa','total_si_cumplio_estado','total_no_cumplio_estado')
        .unique()
        .sort('Entidad_Federativa')
)
```

```python
# Calcular el total general de 'Sí cumplió' y 'No cumplió'
total_si_cumplio_nacl = cumplio_estado["total_si_cumplio_estado"].sum()
total_no_cumplio_nacl = cumplio_estado["total_no_cumplio_estado"].sum()
total_general_nacl = total_si_cumplio_nacl + total_no_cumplio_nacl
```

```python
# Porcentaje 'Sí cumplió' sobre el total general de 'nacl'
porcentaje_si_nacl = total_si_cumplio_nacl / total_general_nacl

# Porcentaje 'No cumplió' sobre el total general de 'nacl'
porcentaje_no_nacl = total_no_cumplio_nacl / total_general_nacl
```

```python
nacl_rows = [
    {
        "Entidad_Federativa": "Nacional",
        "cumplio": "Sí cumplió %",
        "porcentaje": porcentaje_si_nacl
    },
    {
        "Entidad_Federativa": "Nacional",
        "cumplio": "No cumplió %",
        "porcentaje": porcentaje_no_nacl
    }
]
```

```python
# Convertimos la lista de diccionarios a un DataFrame de Polars
df_nacl_rows = pl.DataFrame(nacl_rows)
```

```python
df_long = cumplio_estado.melt(
    id_vars="Entidad_Federativa",
    value_vars=["total_si_cumplio_estado", "total_no_cumplio_estado"],
    variable_name="cumplio_status",
    value_name="conteo"
)
```

```python
df_long = df_long.with_columns(
    pl.col("conteo").sum().over("Entidad_Federativa").alias("total_estado")
).with_columns(
    (pl.col("conteo") / pl.col("total_estado")).alias("porcentaje")
).select(
    pl.col("Entidad_Federativa"),
    # Renombrar la columna 'cumplio_type' para que coincida con el DF 'nacl'
    pl.when(pl.col("cumplio_status") == "total_si_cumplio_estado")
      .then(pl.lit("Sí cumplió %"))
      .otherwise(pl.lit("No cumplió %"))
      .alias("cumplio"),
    pl.col("porcentaje")
)
```

```python
# C. Concatenar el DataFrame transformado con las filas 'nacl'
df_final = pl.concat([df_long, df_nacl_rows], how="vertical")
```

```python
cumplimiento = df_final.rename({
    'cumplio':'Cumplio',
    'porcentaje':'Porcentaje',
})
```

```python
from pandas.api.types import CategoricalDtype

# --- PASO CLAVE: Determinar el Orden del Eje X ---
# Filtrar solo las filas donde 'Cumplio' es "Sí cumplió %"
df_orden = cumplimiento.filter(
    pl.col("Cumplio") == "Sí cumplió %"
)

# Ordenar este subconjunto por 'Porcentaje' de forma ascendente (de menor a mayor)
# Si quieres de mayor a menor, cambia 'ascending=True' a 'ascending=False'
df_orden = df_orden.sort("Porcentaje", descending=False)

# Extraer la lista ordenada de Entidades Federativas
ordered_entities = df_orden["Entidad_Federativa"].to_list()

# --- Aplicar el Orden en el DataFrame (Convertir a Pandas para Plotnine) ---
df_pd = cumplimiento.to_pandas()

# 3.1 Definir el tipo de dato Categorical (Ordinal) con el orden fijo para el EJE X.
entity_cat = CategoricalDtype(categories=ordered_entities, ordered=True)

# 3.2 Aplicar el nuevo tipo de dato a la columna 'Entidad_Federativa'.
df_pd['Entidad_Federativa'] = df_pd['Entidad_Federativa'].astype(entity_cat)

# 3.3 (Opcional pero Recomendado) Aplicar el orden también a la columna 'Cumplio'
# para asegurar que "Sí cumplió %" esté siempre abajo en la pila.
cumplio_order = ["Sí cumplió %", "No cumplió %"]
cumplio_cat = CategoricalDtype(categories=cumplio_order, ordered=True)
df_pd['Cumplio'] = df_pd['Cumplio'].astype(cumplio_cat)
```

```python
#| label: fig-cumplimiento
#| fig-cap: Porcentaje de cumplimiento del Exhorto 02/XLVIII/22, por entidad federativa

from mizani.formatters import percent_format

(
    ggplot(df_pd, aes(x='Entidad_Federativa', y='Porcentaje', fill='Cumplio'))
    + geom_col(position='stack') # 'stack' apila las barras (suma 1.0 por estado)
    # Colores personalizados
    + scale_fill_manual(values={"Sí cumplió %": "#235b4e", "No cumplió %": "#9f2241",})
    + scale_y_continuous(
        breaks=[0,.2,.4,.6,.8,1,],
        limits=(0, 1),
        labels=percent_format(),
    )
    + theme(
        figure_size=(12,8), # width, height
        # customize font source
        #text=element_text(family='Noto Sans'),
        # customize title and subtitle
        plot_title=element_text(size=22, weight='bold', ha='left', margin={'b': 10}, color='#691c32'),
        plot_subtitle=element_text(size=14, weight='bold', ha='left', margin={'b': 10}, color='#bc955c'),
        # customize plot caption
        plot_caption=element_text(size=14, ha='left', va='bottom', color='#98989a', margin={'t': 10}),
        # Customize axis titles
        axis_title_x=element_text(size=14, margin={'t': 10}),
        axis_title_y=element_text(size=12, margin={'r': 10}),
        # Customize axis text (labels)
        axis_text_x=element_text(size=11, angle=80, hjust=1), # Rotate x-axis labels for readability
        axis_text_y=element_text(size=10),
        # remove axes ticks
        axis_ticks_major_x=element_blank(),
        axis_ticks_major_y=element_blank(),
        # Customize legend
        #legend_title=element_text(size=12, color='black', weight='bold', margin={'b': 5}), # Margin below the title
        legend_text=element_text(size=14),
        legend_title=element_blank(),
        legend_position='bottom',
        # Customize panel background and grid lines
        panel_background=element_rect(fill='#fff'), # color background, light gray border
        panel_grid_major=element_text(color='#f5f5f5'), # Subtle major grid lines
        panel_grid_minor=element_text(color='#f5f5f5'), # Even subtler minor grid lines
    )
    # Etiquetas y título
    + labs(
        title='Cumplimiento del 20% del Gasto en Seguridad Pública',
        subtitle='Porcentaje por Entidad Federativa',
        x='',
        y='Porcentaje municipios',
        caption='Fuente: Informes de conclusión FORTAMUN 2024',
    )
).draw()
```

La gráfica presenta, para cada entidad federativa, el porcentaje de municipios que cumplieron con destinar al menos el 20% del FORTAMUN a acciones directamente vinculadas con la seguridad pública.
En el extremo izquierdo se observan entidades como Querétaro, Tamaulipas, Zacatecas, Sonora y Nuevo León, donde el porcentaje de municipios cumplidos es relativamente bajo.
Hacia la parte central aparecen entidades con porcentajes intermedios, como Campeche, Puebla, Colima, Yucatán y Chihuahua.
En el extremo derecho se encuentran diversas entidades con porcentajes elevados de cumplimiento, entre ellas Hidalgo, Aguascalientes, Baja California y Quintana Roo, donde prácticamente la totalidad de los municipios reportaron cumplimiento. 

Más allá del cumplimiento al exhorto, al analizar el ejercicio del gasto del Fondo, se pueden observar contrastes importantes entre estados.
Aguascalientes (74.07%) y Tabasco (65.85%) destinan la mayor proporción del FORTAMUN en seguridad pública.
Les siguen Hidalgo, Baja California, Quintana Roo, Sinaloa y Baja California Sur con porcentajes entre 50% y 60%, lo que indica un compromiso significativo con el fortalecimiento de la seguridad municipal.
En contraste, la mayoría de los estados asigna entre 20% y 40%, mostrando un uso más equilibrado del fondo entre seguridad y otros rubros. Finalmente, entidades como Tamaulipas y Querétaro destinan menos del 10%, priorizando otros gastos municipales por encima de la seguridad.

# Inversión, adquisiciones y beneficiados por subprograma

El ejercicio de los recursos del FORTAMUN destinados a seguridad pública permite identificar cómo los municipios orientaron su gasto hacia acciones alineadas con los Programas con Prioridad Nacional, así como el alcance operativo de dichas intervenciones.
Con base en los Informes de Conclusión remitidos por los municipios, se analizaron 03 elementos:

- Inversión por subprograma
- Adquisiciones realizadas
- Población beneficiada

La calidad de la información varía entre entidades y municipios, por lo que los resultados deben interpretarse considerando dichas diferencias.

## Inversión por subprograma

El ejercicio de los recursos del FORTAMUN destinados a seguridad pública muestra una distribución diferenciada entre los subprogramas alineados a los Programas con Prioridad Nacional.
El análisis del gasto reportado por los municipios evidencia que la mayor parte de los recursos se orientó a fortalecer condiciones laborales y capacidades institucionales.
De acuerdo con los Informes de Conclusión, el subprograma con mayor monto ejercido fue “Mejora salarial, prestaciones y bonos por buen desempeño”, con una inversión de $7,741.95 millones de pesos.
Le siguen “Fortalecimiento de las instituciones de seguridad pública y procuración de justicia” con $3,089.64 millones de pesos, e “Infraestructura y equipamiento de las instituciones de seguridad pública y procuración de justicia” con 536.82 millones de pesos.

```python
subprogramas = conn.sql('''
    WITH subprograma_raw AS (
        SELECT
            UPPER(nom_subprograma) AS subprograma,
            SUM(monto_anual_pagado) AS ejercido
        FROM fortamun_municipios_2024
        GROUP BY 1
    )
    SELECT
        ROW_NUMBER() OVER (ORDER BY subprograma) AS subprograma_id,
        subprograma,
        ejercido/1_000_000 ejercido_mdp
    FROM subprograma_raw
    ORDER BY 3 DESC
''').pl()
```

```python
#| label: fig-gasto-subprogramas
#| fig-cap: Inversión ejercida por subprograma con prioridad nacional

(
    ggplot(subprogramas, aes("reorder(subprograma_id, -ejercido_mdp)", 'ejercido_mdp'))
    # Usar geom_bar con position='dodge' para barras agrupadas
    + geom_bar(stat='identity', position='dodge', width=0.7, fill='#bc955c')
    # Escala para formatear el eje Y en millones
    + scale_y_continuous(limits=(0, 8_000), breaks=range(0, 10_000, 1_000), labels=lambda l: [f"{int(val):,}" for val in l])
    # Ajustes de tema (para simular altura/anchura y estilo)
    + theme(
        figure_size=(12,8), # width, height
        # customize font source
        #text=element_text(family='Noto Sans'),
        # customize title and subtitle
        plot_title=element_text(size=22, weight='bold', ha='left', margin={'b': 10}, color='#691c32'),
        plot_subtitle=element_text(size=14, weight='bold', ha='left', margin={'b': 10}, color='#bc955c'),
        # customize plot caption
        plot_caption=element_text(size=14, ha='left', va='bottom', color='#98989a', margin={'t': 10}),
        # Customize axis titles
        axis_title_x=element_text(size=11, margin={'t': 10}),
        axis_title_y=element_text(size=12, margin={'r': 10}),
        # Customize axis text (labels)
        axis_text_x=element_text(size=10, angle=0, hjust=1), # Rotate x-axis labels for readability
        axis_text_y=element_text(size=11),
        # remove axes ticks
        axis_ticks_major_x=element_blank(),
        axis_ticks_major_y=element_blank(),
        # Customize legend
        #legend_title=element_text(size=12, color='black', weight='bold', margin={'b': 5}), # Margin below the title
        legend_text=element_text(size=10),
        legend_title=element_blank(),
        legend_position='bottom',
        # Customize panel background and grid lines
        panel_background=element_rect(fill='#fff'), # color background, light gray border
        panel_grid_major=element_text(color='#f5f5f5'), # Subtle major grid lines
        panel_grid_minor=element_text(color='#f5f5f5'), # Even subtler minor grid lines
    )
    # Etiquetas y título
    + labs(
        title='Gasto Ejercido por Subprograma',
        subtitle='',
        x='',
        y='Gasto Ejercido (millones de pesos)',
        caption='Fuente: Informes de conclusión FORTAMUN 2024',
    )
)
```

```python
#| label: tbl-subprogramas
#| tbl-cap: Inversión por subprograma con prioridad nacional

(
    GT(subprogramas)
    .tab_header(
        title = "Inversión por subprograma con prioridad nacional",
        subtitle='Importe por entidad federativa'
    )
    .tab_source_note(
        source_note = "Fuente: Informes de conclusión FORTAMUN 2024 "
    )
    .fmt_currency(
        columns="ejercido_mdp", decimals=2
    )
    .cols_label(
        subprograma_id="Id",
        subprograma="Subprograma",
        ejercido_mdp="Importe (mdp)"
    )
    .cols_width(
        subprograma_id="5%",
        subprograma="70%",
        ejercido_mdp="25%",
        )
    .cols_align(
        align="center",
        columns='subprograma_id'
    )
    .opt_table_font(font=["Noto Sans"])
    .tab_options(
        container_width="100%",
        container_height="80%",
        table_width="100%",
        heading_background_color="#691c32",
        column_labels_background_color="#ddc9a3",
        source_notes_background_color="#ddc9a3",
        row_striping_include_table_body = True,
        row_striping_background_color='#f8f8f8',
    )
    .tab_style(
        style=style.text(font="Noto Sans", size="14px"),
        locations=loc.body(),
    )
    .opt_vertical_padding(scale=0.3)
    .opt_horizontal_padding(scale=0.4)
)
```

El comportamiento del gasto muestra una concentración clara en acciones relacionadas con remuneraciones y condiciones laborales, seguidas por la inversión en fortalecimiento institucional y equipamiento, mientras que los subprogramas de capacitación, prevención, videovigilancia o atención de llamadas de emergencia presentan montos menores.
Este patrón sugiere que los municipios destinaron la mayor parte de sus recursos a atender necesidades estructurales y operativas inmediatas, principalmente asociadas a compromisos salariales y adquisición de equipo básico.

## Adquisiciones por subprograma 

La siguiente gráfica muestra la distribución de las principales adquisiciones realizadas por los distintos subprogramas, destacando las partidas genéricas con mayor inversión. # Estado de fuerza

El estado de fuerza municipal se construyó a partir del número de elementos de policía reportados por cada municipio en su Informe de Conclusión, combinado con la población municipal estimada, para obtener la tasa de policías por cada mil habitantes.
De acuerdo con los Criterios FORTAMUN, todos los informes deben contener esta información; sin embargo, varios municipios no requisitaron el campo correspondiente o lo dejaron en blanco.
Esto genera incertidumbre sobre si se trata de municipios sin cuerpo policial o simplemente de omisiones en el llenado.

```python
edo_fza_mun = conn.sql('''
    WITH edo_fza_raw AS (
        SELECT
            entidad_federativa,
            municipio,
            SUM(edo_fza) edo_fza,
        FROM fortamun_municipios_2024
        GROUP BY 1, 2
    )
    SELECT
        ROW_NUMBER() OVER (ORDER BY municipio) AS municipio_id,
        entidad_federativa,
        municipio,
        edo_fza,
    FROM edo_fza_raw
    ORDER BY 2, 3
''').pl().with_columns(
        pl.col('edo_fza').cast(pl.Int32)
    )
```

```python
pob_mun = (
    pl.read_excel('pob_mun.xlsx')
        .with_columns(pl.col('NOM_MUN').str.to_uppercase())
)
```

```python
edo_fza_2 = (
    edo_fza_mun
        .join(pob_mun, left_on='municipio', right_on='NOM_MUN', how='inner')
        .select('entidad_federativa','municipio', 'poblacion','edo_fza')
        .with_columns(
            (pl.col('edo_fza')/pl.col('poblacion')*1000).round(4).alias('tasa_pol')
        )
)
```

```python
df_edo_fza = edo_fza_2.to_pandas()
```

```python
median_order = (
    df_edo_fza.groupby('entidad_federativa')['tasa_pol']
    .median()
    .sort_values(ascending=False)
    .index.tolist()
)
```

```python
df_edo_fza['entidad_federativa'] = pd.Categorical(
    df_edo_fza['entidad_federativa'],
    categories=median_order,
    ordered=True
)
```

# Retenciones 

Las retenciones aplicadas a las ministraciones del FORTAMUN corresponden, en términos generales, a descuentos realizados por los gobiernos estatales para cubrir adeudos municipales, particularmente con organismos operadores de agua, servicios locales u otros compromisos financieros.
Estas retenciones reducen el monto efectivamente recibido por los municipios y, por lo tanto, pueden incidir en la disponibilidad de recursos destinados a la atención de necesidades de seguridad pública. 

Para el ejercicio 2024, los municipios reportaron un total de $195,190,000 pesos en retenciones aplicadas por CONAGUA.

Entre los municipios que sí registraron afectaciones, la retención promedio asciende a $2,693,333 pesos.

Los montos promedio muestran una variabilidad importante entre entidades federativas.
